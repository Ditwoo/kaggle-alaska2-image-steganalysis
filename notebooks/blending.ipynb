```python
dfs = [ps.read_csv(file).sort_values("Id") for file in data]

dfs[0].head()
```

```python
mask = np.zeros((len(dfs), len(dfs)), dtype=np.bool) 
# mask[np.triu_indices_from(mask)] = True
# mask[np.diag_indices_from(mask)] = False

corr_matr = np.zeros((len(dfs), len(dfs)))

for i in range(len(dfs)):
    for j in range(len(dfs)):
        a = dfs[i]["Label"].values
        b = dfs[j]["Label"].values
        corr_matr[i, j] = stats.spearmanr(a, b).correlation
    
# corr_matr = np.tril(corr_matr)
files = [f.rsplit("/")[-1] for f in data]

plt.figure(figsize=(10, 10))
sbn.heatmap(
    corr_matr,
    mask=mask,
    annot=True,
    xticklabels=files,
    yticklabels=files
)
plt.title("submits correlation");
```

```python
def blend(values, weights = None):
    if weights is not None:
        if len(values) != len(weights):
            raise ValueError(
                "Number of values should be the same as number of weights!"
            )
        if np.sum(weights) != 1:
            raise ValueError(
                "Sum of the weights should be equal to 1!"
            )
        res = np.mean(np.stack([vec * w for vec, w in zip(values, weights)]), 0)
    else:
        res = np.mean(np.stack([vec for vec in values]), 0)
    return res
```

```python
def power_avg(values, power: int = 1):
    return np.mean(np.stack([vec ** power for vec in values]), 0)
```

```python
def rank_avg(values):
    res = np.zeros(values[0].shape[0], dtype=np.float)
    for vec in values:
        res = np.add(res, stats.rankdata(vec) / vec.shape[0])
    res = res / len(values)
    return res
```
